```python
#BAR PLOT
# switch female to left and male to right
sns.barplot(x="Gender" , y="Age (years)-Float/Int" , hue="Marital Status?" , data=df , order=["Female", "Male"])
plt.title(" Title ")
plt.show()
```

```python
#BAR PLOT
# Color Pelets
sns.barplot(x="Gender" , y="Age (years)-Float/Int" , hue="Marital Status?" , data=df , order=["Female", "Male"] , color="gray")
plt.title(" Title ")
plt.show()
```

```python
#BAR PLOT
# Color Pelets
# ci = none removes or stick the arrow on top of bar
# and pallete gives the sea born pallet some color we can find different pallets inside seaborn from google
sns.barplot(x="Gender" , y="Age (years)-Float/Int" , hue="Marital Status?" , data=df , order=["Female", "Male"] , color="brown" ,
            palette = 'pastel' ,  ci=None)
plt.title(" Title ")
plt.show()
```

```python
#Horizontal Plot by taking string variable on y axis and numeric variable on x asis
#BAR PLOT

from numpy import mean
sns.barplot(x="Age (years)-Float/Int" , y="Gender" , hue="Marital Status?" , data=df , order=["Male", "Female"] , color="brown" ,
            palette = 'pastel',estimator=mean , saturation=0.2)
plt.title(" Title ")
plt.show()
```

```python
# now instead of palette entering our own values
# line width changes the motai of line
# errcolor err is actually this arrow in graph in centre of bar % 
# err & edge color value ranges from 0.0 to 0.9
# here in facecolor 4 values are RGB is red green blue values in an image and last is Alpha which is value for transparent or opaque intensity
# first RGB 3 values combine to give a color and then in last Alpha give values for opaque intensity
sns.barplot(x="Gender" , y="Age (years)-Float/Int", data=df ,
            linewidth=3, facecolor=(0.5,0.1,0.3,0.9),
             errcolor="0.2", edgecolor="0.5")
plt.title(" Title ")
plt.show()
```
