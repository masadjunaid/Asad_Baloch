## Assignment 1 

```python
import numpy as np
import pandas as pd 
import seaborn as sns

df = sns.load_dataset('titanic')

df.isnull().sum()
df.head()

```

```python
# our assignment is to replace mean values of deck 
# as deck  we will have to remove the entire column as we cannot replace values with mean
df = df.drop('deck', 1)
df.head()
```

```python
df.isnull().sum()
# we can see as follows deck column is gone as it had most missing values

```

```python
# following will remove missing values in embarked, and embarked town
df.dropna(subset=['embarked' ], axis=0 , inplace=True)
df.isnull().sum()
```

# Assignment 2 

```python
# convert age to days
df['age'] = df['age']*365
df.head(10)
```

```python
# another assignment is to remove zeros we can easily do it by converting to int
# but we will first have to remove NaN of age , otherwise it wont convert NaN to int and give error 
df.dropna(subset=['age' ], axis=0 , inplace=True)
df['age'] = df['age'].astype("int64")
df.head(10)
```

```python
# Find mode of multiple columns in one code or we can also use mode method we will see this also
df1 = sns.load_dataset('titanic')




df1_multi = df1.filter(['deck', 'embarked', 'embark_town']).mode()

df1_multi



# Create a sub-set and replace nan with mode in one code in all three columns.

cols = ['deck', 'embarked','embark_town']

df1[cols] = df1[cols].fillna(df1.mode().iloc[0])
df1.isnull().sum()

# result is showing clearly all NaN values are replaced by Mode which is most repetative value in categorical variable 
```
